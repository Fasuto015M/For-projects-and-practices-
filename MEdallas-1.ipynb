```python
df2= df.copy()
df2.sample(6)
```

```python
#Reemplazando los datos nulos por los datos promedio.
for col in ['Medallas_Oro', 'Medallas_Plata', 'Medallas_Bronce']:
    df2[col] = df2[col].fillna(
        df2.groupby('Pais')[col].transform('mean')
    )
```

```python
# Agrupar por país
medallas_pais = df2.groupby('Pais')['Medallas_Oro','Medallas_Plata','Medallas_Bronce'].sum().sort_values(by= 'Pais', ascending=False)

# Graficar
plt.figure(figsize=(10,6))
medallas_pais.plot(kind='bar', edgecolor='black')

plt.title("Total de Medallas de Oro por País")
plt.xlabel("País")
plt.ylabel("Cantidad de Medallas de Oro")
plt.xticks(rotation=45)
plt.show()
```

```python
# Agrupar por país
medallas_pais = df2.groupby('Pais')['Medallas_Oro'].sum().sort_values(ascending=False)

# Graficar
plt.figure(figsize=(10,6))
medallas_pais.plot(kind='bar', edgecolor='black')

plt.title("Total de Medallas de Oro por País")
plt.xlabel("País")
plt.ylabel("Cantidad de Medallas de Oro")
plt.xticks(rotation=45)
plt.show()
```

```python
# Agrupar por país
medallas_pais = df2.groupby('Deporte')['Medallas_Oro','Medallas_Plata','Medallas_Bronce'].sum().sort_values(by= 'Deporte', ascending=False)

# Graficar
plt.figure(figsize=(10,6))
medallas_pais.plot(kind='bar', edgecolor='black')

plt.title("Total de Medallas por Deporte")
plt.xlabel("País")
plt.ylabel("Cantidad de Medallas por Deporte")
plt.xticks(rotation=45)
plt.show()
```

```python
# Agrupar por país
medallas_pais = df2.groupby('Anio')['Medallas_Oro','Medallas_Plata','Medallas_Bronce'].sum().sort_values(by= 'Anio', ascending=False)

# Graficar
plt.figure(figsize=(10,6))
medallas_pais.plot(kind='bar', edgecolor='black')

plt.title("Total de Medallas por año")
plt.xlabel("País")
plt.ylabel("Cantidad de Medallas por año")
plt.xticks(rotation=45)
plt.show()
```
